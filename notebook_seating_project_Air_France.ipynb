# <font color=red>**Modélisation**</font>

We changed how we wanted to resolve the problem: instead of trying to optimize the best seat layout in one go. We first try to optimize and bring people in groups on
    the same row. Then on a second run we try to optimize and optimize on the row level so people can sit next to each other.

The idea is to first resolve the problem on the row level with every constraint function in the model as an objective function to satisphy most passengers to bring them together on the rows (the best optimization is each member of a group are on the same row) and the number of passengers 
    that are in transit sit in front of the plane

In the second run, we try to resolve a second model, by removing some constraints (for example move the passengers that are of the same class type ), but we are adding some for example to not change the row given to a passenger after the first optimization run.

## <font color=blue>Variables</font>
Depuis le premier jalon, la modélisation a été changée pour améliorer le temps de calcul de notre modèle : nous considérons maintenant des variables binaires dans une matrice :
$$M = (x_{ij})_{i\in passagers,\ j\in sièges}$$
$$x_{ij} =\left(\begin{matrix} 1\ si\ le\ passager\ i\ est\ à\ la\ place\ j \\0\ sinon\end{matrix}\right.$$

## <font color=blue>Contraintes</font>
### <font color=peru>Unicité de l'attribution des sièges et des passagers</font>
Ainsi, pour les contraintes de l'unicité de l'attribution d'un siège à un passager, il suffit que la somme des coefficients de la ligne du passager soit égal à 1 (exactement un siège lui a été attribué).

De même pour l'attribution d'un passager à un siège, il faut que la somme des coefficients de la colonne du siège soit inférieur ou égal à 1 (on lui attribue un passager ou aucun s'il y a plus de passager que de sièges).

### <font color = peru>Gestion des classes</font>

Pour la gestion des classes économique et business, la méthode choisie est d'imposer les contraintes suivantes : Le nombre de passagers avec un ticket business se retrouvant en classe économique doit être nul", et "Le nombre de passagers avec un ticket économique se retrouvant en classe business doit être nul".

Pour ce faire, nous avons choisi de représenter cela par un calcul matriciel. On introduit un vecteur pour filtrer les sièges, un autre pour filtrer les passagers et on multiplie de part et d'autre la matrice $M$ contenant les variables de décisions. On introduit le vecteur $Ticket_B$ qui représente les voyageurs avec un ticket business :
$$Ticket_B = (Ticket_{B_i}),\ où\ Ticket_{B_i} = 1\ si\ le\ passager\ i\ a\ un\ ticket\ business,\ 0\ sinon$$

De même, on introduit le vecteur $Place_{E}$ qui représente les places en classe économique :

$$Place_E = (Place_{E_j}),\ où\ Place_{E_i} = 1\ si\ le\ siège\ j\ est\ en\ classe\ économique,\ 0\ sinon$$

Le produit $M\times Place_E$ représente une matrice dont le coefficient en position $(i,j)$ vaut 1 si le passager $i$ a été attribué au siège $j$ et que celui_ci est en classe économique.

Ainsi, le produitscalaire $Ticket_B^\top\cdot (M\times Place_E)$ représente le nombre de voyageurs ayant un ticket pour une classe business ayant été affecté à un siège en classe économique : la contrainte associée est la suivante :
$$Ticket_B^\top\cdot (M\times Place_E) <= 0$$

On fait pareil pour le la situation opposée en remplaçant les 0 par des 1 et les 1 par des 0 dans les vecteurs $Ticket_B$ et $Place_E$.

### <font color = peru>Voyageurs en transit</font>
Pour les voyageurs en transit, on calcule le nombre $r_{\frac{1}{3}}$ de voyageurs en transit dans le premier tiers de chaque classe. Pour cela, on utilise deux vecteurs : 
$$Transit = \left(transit_i\right),\ où\ transit_i = 1\ si\ le\ passager\ i\ a\ un\ transit\ court,\ 0\ sinon$$

$$Avant = \left(avant_j\right),\ où\ avant_j = 1\ si\ le\ siège\ j\ est\ dans\ le\ premier\ tiers\ de\ sa\ classe,\ 0\ sinon$$

L'objectif serait alors de maximiser le produit scalaire, qui correspond au nombre de personnes en transit rapide (moins de 90 minutes) dont le siège est dans le premier tiers de leur classe :
$$r_\frac{1}{3} = Transit^\top\cdot (M\times Avant)$$

### <font color = peru>Gestion des enfants</font>
Pour imposer que les enfants aient à côté un adulte, on impose:

$$\forall i \in [1;Nombre\_ places],  (Mat_{child,i} = 1 \implies \sum_{place \in voisins(i)} \sum_{adulte \in groupe} Mat_{adulte,place} \geq 1) $$

avec voisins les indices des places voisines de la place i

### <font color = peru>Gestion des personnes à mobilité réduite</font>

Pour imposer que les personnes handicapées en chaise roulante (WCHR) soient situées selon les contraintes, on impose qu'elles soient sur une place adjacente à l'allée centrale et on crée pour chaque WCHR trois personnes fictives de masse nulle que l'on attribue sur les places que l'on souhaite laisser libres. On appellera par la suite les personnes fictives par l'appellation "fantome"

Pour imposer que les personnnes en brancard (WCHB) soient situées selon les contraintes, on impose qu'elles soient placées dans un milieu de rangée et on empêche quiconque de s'installer sur les trois places du rang de devant, du rang de l'individu et des deux rangs de derrière

## <font color = blue>Fonction objectif pour le premier modèle</font>

L'objectif de cette première optimisation est de correctement placer les personnes en transit ainsi que de mettre les groupes sur un même rangée.
Pour cela, on considère la fonction objectif suivante :
$$min \sum_{g \in groupes}{(max\ \underset{i,j\in g}{dist_{rangee}}(i,j))}- r_\frac{1}{3}$$

## <font color = blue>Fonction objectif pour le deuxième modèle</font>

Dans ce deuxième temps, les voyageurs en transit ont déjà été placés et on ne cherche plus à optimiser selon les rangées puisqu'on ne peut plus changer la rangée attribuée à un passager. On aura alors la fonction objectif suivante :
$$min \sum_{g \in groupes}{(max\ \underset{i,j\in g}{dist_{siege}}(i,j))}$$

### Extraction des données de l'Excel

```python
def extractExcel(filename):
    """Remplit un dictionnaire avec les informations du fichier en entrée"""
    dataframe = pd.ExcelFile(filename)
    sheet_dict = {}
    i = 0
    for sheet_name in dataframe.sheet_names:
        sheet_dict.update({i: dataframe.parse(sheet_name)})
        i = i + 1
    return sheet_dict

def extractMinuteFromTransitTime(transitTime):
    minutesInHour = 60
    
    time = 0
    time = int(transitTime.hour * minutesInHour + transitTime.minute + transitTime.second / minutesInHour)
    
    return time

def createDicOfGroups(sheet_dict):
    """
    Renvoie un dict où pour chaque groupe, on a le nombre de personnes dans chaque catégorie,
    le temps de transit et la classe
    """
    final_sheet = {}

    for i in range(0, len(sheet_dict)):
        tempmale = 0
        tempfemale = 0
        tempnrgroupe = 0
        time = 0
        tempchild = 0
        tempWCHR = 0
        tempWCHB = 0
        tempGhost = 0
        Class = 0


        if sheet_dict['Numéro du groupe'][i].is_integer():
            tempnrgroupe = sheet_dict['Numéro du groupe'][i]

        if  sheet_dict['Hommes'][i].is_integer() and tempnrgroupe != 0:
            tempmale = sheet_dict['Hommes'][i]

        if sheet_dict['Femmes'][i].is_integer() and tempnrgroupe != 0:
            tempfemale = sheet_dict['Femmes'][i]

        if sheet_dict['Enfants'][i].is_integer() and tempnrgroupe != 0:
            tempchild = sheet_dict['Enfants'][i]

        if sheet_dict['WCHR'][i].is_integer() and tempnrgroupe != 0:
            tempWCHR = sheet_dict['WCHR'][i]
            tempGhost = 3

        if sheet_dict['WCHB'][i].is_integer() and tempnrgroupe != 0:
            tempWCHB = sheet_dict['WCHB'][i]

        if not isinstance(sheet_dict['TransitTime'][i], float):
            temptime = sheet_dict['TransitTime'][i]
            time = extractMinuteFromTransitTime(temptime)
        
        if isinstance(sheet_dict['Classe'][i], str): # Business class = 1, Economy class = 0
            tempclass = sheet_dict['Classe'][i]
            if tempclass == "J":
                Class = 1
            elif tempclass == "Y":
                Class = 0

        if tempnrgroupe != 0:
            final_sheet.update({int(tempnrgroupe-1) : 
                                {'h' : int(tempmale), 'f' : int(tempfemale), 'c' : int(tempchild), 
                                 'WCHR' : int(tempWCHR), 
                                 'WCHB' : int(tempWCHB), 'g': int(tempGhost), 
                                 'transittime' : time, 'class' : Class}})

    return final_sheet

def createDicOfPassengers(sheet_dict):
    """
    Retourne un dictionnaire où pour chaque passager, on a :
    sa catégorie, son temps de transit, son numéro de groupe et sa classe
    """
    final_sheet = {}

    index = 0
    Class = 0
    for i in range(0, len(sheet_dict)): # On itère sur chaque groupe dans le tableau excel
        tempnrgroupe = 0 # On stocke le numéro de groupe d'un passager

        if sheet_dict['Numéro du groupe'][i].is_integer(): 
            tempnrgroupe = sheet_dict['Numéro du groupe'][i]

        if not isinstance(sheet_dict['TransitTime'][i], float):
            temptime = sheet_dict['TransitTime'][i]
            time = extractMinuteFromTransitTime(temptime)
        
        if isinstance(sheet_dict['Classe'][i], str):
            tempclass = sheet_dict['Classe'][i]
            if tempclass == "J":
                Class = 1
            elif tempclass == "Y":
                Class = 0

        if  sheet_dict['Hommes'][i].is_integer() and tempnrgroupe != 0:
            tempmale = sheet_dict['Hommes'][i]
            for j in range(0, int(tempmale)): # On stocke les infos de tous les hommes du groupe
                final_sheet.update({index: {'cat': 'h', 'transittime' : time, 
                                            'groupnr' : int(tempnrgroupe-1), 'class' : Class}})
                index += 1

        if sheet_dict['Femmes'][i].is_integer() and tempnrgroupe != 0:
            tempfemale = sheet_dict['Femmes'][i]
            for j in range(0, int(tempfemale)): # On stocke les infos de toutes les femmes du groupe
                final_sheet.update({index: {'cat': 'f', 'transittime' : time, 
                                            'groupnr' : int(tempnrgroupe-1), 'class' : Class}})
                index += 1

        if sheet_dict['Enfants'][i].is_integer() and tempnrgroupe != 0:
            tempchild = sheet_dict['Enfants'][i]
            for j in range(0, int(tempchild)):# On stocke les infos de tous les enfants du groupe
                final_sheet.update({index: {'cat': 'e', 'transittime' : time, 
                                            'groupnr' : int(tempnrgroupe-1), 'class' : Class}})
                index += 1
        
        if sheet_dict['WCHR'][i].is_integer() and tempnrgroupe != 0:
            tempWCHR = sheet_dict['WCHR'][i]
            for j in range(0, int(tempWCHR)):# On stocke les infos de tous les WCHR du groupe
                final_sheet.update({index: {'cat': 'WCHR', 'transittime' : time, 
                                            'groupnr' : int(tempnrgroupe-1), 'class' : Class}})
                index += 1
                for i in range(0, 3): # category ghost (un ghost est un passager fantôme autour d'un WCHR)
                    final_sheet.update({index: {'cat' : 'g', 'transittime' : 0, 
                                            'groupnr' : int(tempnrgroupe-1), 'class' : Class}})
                    index += 1
        
        if sheet_dict['WCHB'][i].is_integer() and tempnrgroupe != 0:
            tempWCHB = sheet_dict['WCHB'][i]
            for j in range(0, int(tempWCHB)):# On stocke les infos de tous les WCHB du groupe
                final_sheet.update({index: {'cat': 'WCHB', 'transittime' : time, 
                                            'groupnr' : int(tempnrgroupe-1), 'class' : Class}})
                index += 1

    return final_sheet


def masse(index, passenger_sheet):
    """
    Retourne la masse du passager selon sa catégorie (inclue les passagers fantômes)
    """
    if passenger_sheet[index]['cat']=='h':
        return 90
    if passenger_sheet[index]['cat']=='f':
        return 70
    if passenger_sheet[index]['cat']=='e':
        return 50
    if passenger_sheet[index]['cat']=='WCHR':
        return 15 + 80
    if passenger_sheet[index]['cat']=='WCHB':
        return 80
    if passenger_sheet[index]['cat']=='g':
        return 0
```

```python
def length_of_group(g):
    amount = g['h'] + g['f'] + g['c'] + g['WCHR'] + g['WCHB']
    return amount


def obtain_vector_passenger_places(NumberOfSeats,NumberOfBusinessSeats):
    n=1
    vector=np.zeros(NumberOfSeats,dtype=int)
    for i in range (NumberOfSeats):
        if i < NumberOfBusinessSeats:
            vector[i] = n
            if n == 5:
                n=1
            elif n==2:
                n=4
            else:
                n+=1
        else:
            vector[i]=n
            if n==7:
                n=1
            elif n==3:
                n=5
            else:
                n+=1
    return vector

def obtain_vector_passenger_places_bary(NumberOfSeats,NumberOfBusinessSeats):
    n=1
    vector=np.zeros(NumberOfSeats,dtype=int)
    for i in range (NumberOfSeats):
        if i < NumberOfBusinessSeats:
            vector[i] = n
            if n == 7:
                n=1
            else:
                n+=2
        else:
            vector[i]=n
            if n==7:
                n=1
            elif n==3:
                n=5
            else:
                n+=1
    return vector

def obtain_vector_passenger_rows(NumberOfSeats,NumberOfBusinessSeats):
    n=1
    row=1
    vector=np.zeros(NumberOfSeats,dtype=int)
    for i in range (NumberOfSeats):
        if i < NumberOfBusinessSeats:
            vector[i]=row
            n+=1
            if n==5:
                n=1
                row+=1
        else:
            vector[i]=row
            n+=1
            if n==7:
                n=1
                row+=1
    return vector

def place_to_paire(place):
    """
    Renvoie le couple des places associés au siège de numéro place
    Fonction utilisé pour l'affichage à la fin
    """
    if place < NumberOfBusinessSeats:
        a,b = place//4 + 1, place % 4 + 1
        b = 2*b -1
    else:
        place -= NumberOfBusinessSeats
        a,b = place//PlacesDansRang + 1 + NumberOfBusinessSeats//4, place % PlacesDansRang + 1
        if b >= 4:
            b += 1    
    return (a,b)




#Transit functions
def obtain_vector_passenger_front(NumberOfSeats,NumberOfBusinessSeats):
    n=1
    vector=np.zeros(NumberOfSeats,dtype=int)
    EcoSeats = NumberOfSeats - NumberOfBusinessSeats
    for i in range(((NumberOfBusinessSeats//4)//3)*4):
        vector[i] = 1
    for i in range(((EcoSeats//PlacesDansRang)//3)*PlacesDansRang):
        vector[NumberOfBusinessSeats + i] = 1
    return vector

def obtain_transit_vector(passenger_sheet):
    vector = np.zeros(len(passenger_sheet),dtype=int)
    for i in range(len(vector)):
        if passenger_sheet[i]['transittime'] <= 90 and passenger_sheet[i]['transittime'] > 0:
            vector[i] = 1
    return vector

#Classes functions
def get_number_of_business_seats(passenger_sheet):
    NumberOfBusinessSeats = 0
    for i in range(len(passenger_sheet)):
        if passenger_sheet[i]['class']:
            NumberOfBusinessSeats += 1
    return (NumberOfBusinessSeats//4 + 1) * 4

def obtain_class_vector(passenger_sheet):
    vector = np.zeros(len(passenger_sheet),dtype = int)
    for i in range(len(vector)):
        if passenger_sheet[i]['class']:
            vector[i] = 1
    return vector

def obtain_eco_class_vector(NumberOfSeats,NumberOfBusinessSeats):
    vector = np.zeros(NumberOfSeats,dtype = int)
    for i in range(NumberOfBusinessSeats,NumberOfSeats):
        vector[i] = 1
    return vector
```

### Définition du modèle et des variables

```python
NbPassengers = len(list(passenger_sheet))
NumberOfBusinessSeats = get_number_of_business_seats(passenger_sheet)
NumberOfSeats = NumberOfBusinessSeats + (NombreRangs - NumberOfBusinessSeats//4) * PlacesDansRang

m = Model("Modele 1")

rho = {g:m.addVar(name = f"rho_{g}") for g in range(len(group_sheet))}

mat=m.addMVar((NbPassengers,NumberOfSeats),vtype=GRB.BINARY)
```

### Contraintes d'unicité de place

```python
vector_places=obtain_vector_passenger_places(NumberOfSeats,NumberOfBusinessSeats)
vector_rows=obtain_vector_passenger_rows(NumberOfSeats,NumberOfBusinessSeats)
vector_places_bary = obtain_vector_passenger_places_bary(NumberOfSeats,NumberOfBusinessSeats)

def fplace(id_passenger):
    return vector_places@mat[id_passenger]
    

def frang(id_passenger):
    return vector_rows@mat[id_passenger]

def fplacebary(id_passenger):
    return vector_places_bary@mat[id_passenger]


"""Contrainte: unicité sièges/passagers"""
for i in range(NbPassengers):
    m.addConstr(quicksum(mat[i]) == 1)
for j in range(NumberOfSeats):
    m.addConstr(quicksum(mat[i][j] for i in range(NbPassengers)) <= 1)
```

### Gestion du transit

```python
vector_front = obtain_vector_passenger_front(NumberOfSeats,NumberOfBusinessSeats)
transit_vector = obtain_transit_vector(passenger_sheet)

def transit_front():
    return (transit_vector@mat)@vector_front

r1_3 = transit_front()
```

### Contraintes des classes

```python
class_vector = obtain_class_vector(passenger_sheet)
eco_vector = obtain_eco_class_vector(NumberOfSeats,NumberOfBusinessSeats)

def bad_placed_business():
    return class_vector@mat@eco_vector

def bad_placed_eco():
    opposed_class_vector = np.ones(len(passenger_sheet),dtype = int) - class_vector
    business_vector = np.ones(NumberOfSeats,dtype = int) - eco_vector
    return opposed_class_vector@mat@business_vector

m.addConstr(bad_placed_business() <= 0)
m.addConstr(bad_placed_eco() <= 0)
```

```python
def voisins(id_siege):
    """Renvoie la liste des indices des sièges immédiatement à droite/à gauche selon l'existence"""
    place = vector_places[id_siege]

    if place == 1:
        return [id_siege + 1]
    elif place == 7:
        return [id_siege - 1]
    else:
        return [id_siege - 1, id_siege + 1]

def voisins_rangs(id_siege):
    return [siege for siege in range(len(vector_rows)) if vector_rows[siege]==vector_rows[id_siege]]

def voisins_fantomes(id_siege):
    """Renvoie la liste des sièges où sont installés les passagers fictifs autour d'une chaise roulante"""
    place = vector_places[id_siege]

    if place == 3:
        return [id_siege-PlacesDansRang-1,id_siege-PlacesDansRang,id_siege-1]
    elif place == 5:
        return [id_siege - PlacesDansRang,id_siege-PlacesDansRang+1,id_siege+1]
    else:
        return []

def voisins_civieres(id_siege):
    """"Renvoie la liste des places laissées vides par la présence d'une civière à l'indice id_siege"""
    rang = vector_rows[id_siege]
    place = vector_places[id_siege]
    if rang >= NombreRangs - 2 or rang <= NumberOfBusinessSeats//4 + 1 or not place in [2,PlacesDansRang]:
        return []
    else:
        return [id_siege - PlacesDansRang-1, id_siege-PlacesDansRang,id_siege-PlacesDansRang+1,id_siege-1,id_siege+1,id_siege+PlacesDansRang-1,id_siege+PlacesDansRang,id_siege+PlacesDansRang+1,id_siege+2*PlacesDansRang-1,id_siege+2*PlacesDansRang,id_siege+2*PlacesDansRang+1]

def ischild(index, passenger_sheet):
    return passenger_sheet[index]['cat']=='e'

def isWCHR(index, passenger_sheet):
    return passenger_sheet[index]['cat']=='WCHR'

def isfantome(index, passenger_sheet):
    return passenger_sheet[index]['cat']=='g'

def isWCHB(index, passenger_sheet):
    return passenger_sheet[index]['cat']=='WCHB' 
```

### Contraintes des enfants et passagers à mobilité réduite

```python
parents = {}                        #dictionnaire avec indice = identifiant d'un enfant et valeur = liste des adultes du groupes
fantomes = {}                       #dictionnaire avec indice = identifiant d'un WCHR et valeur = liste des passagers fictifs du groupe
civieres = []                       #liste des passagers en civière
groups = {}                         #dictionnaire avec indice = indice du groupe et valeur = (dictionnaire avec indice = identifiant d'un individu du groupe et valeur = caractéristiques de l'individu)

for child in passenger_sheet:
    if ischild(child,passenger_sheet):
        for id_siege in range(NumberOfSeats):
                if vector_rows[id_siege] in [11,12]:       #pas à côté d'une sortie de secours
                    m.addConstr(mat[child,id_siege] <= 0)

for group in group_sheet:
    groups[group] = {}
    for passenger in passenger_sheet:
           if passenger_sheet[passenger]['groupnr'] == group:
                groups[group][passenger] = passenger_sheet[passenger]

for group in group_sheet:
    for individual in groups[group]:
        if ischild(individual, passenger_sheet) and len([person for person in groups[group] if not isWCHB(person, passenger_sheet)]) != 1:
            if len([child for child in groups[group] if ischild(child, passenger_sheet)]) <= 2*len([adult for adult in groups[group] if not ischild(adult, passenger_sheet)]):
                parents[individual] = [adult for adult in groups[group] if not ischild(adult, passenger_sheet)]
            else:
                parents[individual] = [person for person in groups[group] if person != individual]

        if isWCHR(individual,passenger_sheet):
            fantomes[individual] = [fantome for fantome in groups[group] if isfantome(fantome, passenger_sheet)]
        if isWCHB(individual,passenger_sheet):
            civieres.append(individual)


for child in parents:                               #on impose dans cette boucle qu'un adulte au moins se situe à côté de chaque enfant
    for id_siege in range(NumberOfSeats):
        m.addConstr(quicksum([quicksum([mat[parent, voisin] for voisin in voisins_rangs(id_siege)]) for parent in parents[child]]) >= mat[child,id_siege])

for handicape in fantomes:                          #on impose ici la position des passagers fictifs autour d'un WCHR et le positionnement côté allée centrale de ceux-ci
    for id_siege in range(NumberOfSeats):
        if not vector_places[id_siege] in [3,5]:
            m.addConstr(mat[handicape,id_siege] <= 0)
        else:
            m.addConstr(quicksum([quicksum([mat[fantome, voisin] for voisin in voisins_fantomes(id_siege)]) for fantome in fantomes[handicape]]) >= 3*mat[handicape,id_siege])

for WCHB in civieres:                               #on impose ici de libérer les quatre rangées correspondantes à une WCHB
    for pos_WCHB in range(NumberOfSeats):
        rang = vector_rows[pos_WCHB]
        place = vector_places[pos_WCHB]
        if rang >= NombreRangs - 2 or rang <= NumberOfBusinessSeats//4 + 1 or place not in [2,PlacesDansRang]:
            m.addConstr(mat[WCHB,pos_WCHB] <= 0) 
        else:
            for passager in range(NbPassengers):
                for id_siege in voisins_civieres(pos_WCHB):
                    m.addConstr(1-mat[passager,id_siege] >= mat[WCHB,pos_WCHB])
            for WCHB2 in civieres:
                m.addConstr(1-mat[WCHB2, pos_WCHB + 3*PlacesDansRang] >= mat[WCHB, pos_WCHB])
```

### Approche heuristique

```python
import random
"""
arguments : 
passenger_sheet
group_dict
attributions
"""

def score_groupe(group_nb, group_dict, dict_attributions):
    """Calcule le score d'un groupe (le but final étant de minimiser la somme des scores)"""
    score = 0
    n = len(group_dict[group_nb])
    for i in range(n):
        for j in range(i+1,n):
            rgi, pli = dict_attributions[group_dict[group_nb][i]]
            rgj, plj = dict_attributions[group_dict[group_nb][j]]
            if rgi==rgj:
                score += 30* (abs(pli-plj)-2)
            else:
                score += (50*abs(rgi-rgj) + abs(pli-plj))
    return score

def create_group_scores(group_dict, dict_attributions):
    """Calcule les scores de tous les groupes"""
    scores = {}
    total_score = 0
    for i in group_dict.keys():
        s=score_groupe(i, group_dict, dict_attributions)
        total_score+=s
        scores[i]=s
    scores[-1]=total_score
    return scores

def has_civ(passenger):
    gr_num = passenger_sheet[passenger]['groupnr']
    for i in groups[gr_num]:
        if passenger_sheet[i]['cat'] == "WCHB":
            return True
    return False
        

def candidates(passenger_sheet, group_dict):
    """Renvoie les indices des passagers adultes et en classe business, puis ceux dans un groupe d'au moins 2 personnes"""
    candidates=[[],[]]
    for i in passenger_sheet.keys():
        if passenger_sheet[i]['cat'] in ['h', 'f'] and passenger_sheet[i]['class']==0:
            candidates[0].append(i)
            if len(group_dict[passenger_sheet[i]['groupnr']])>=2:
                candidates[1].append(i)
    return candidates



def barycentre(attributions):
    rg = sum([attributions[i][0]*masse(i, passenger_sheet) for i in attributions.keys()])/sum([masse(i, passenger_sheet) for i in attributions.keys()])
    pl = sum([attributions[i][1]*masse(i, passenger_sheet) for i in attributions.keys()])/sum([masse(i, passenger_sheet) for i in attributions.keys()])
    return (rg,pl)

def contrainte_barycentre(attributions):
    rg,pl = barycentre(attributions)
    return(Bornes['inf']['Rang']<=rg and Bornes['sup']['Rang'] >= rg and Bornes['inf']['Place']<=pl and Bornes['sup']['Place']>=pl)

places_vides_b = {i: {j:True for j in range(1,PlacesDansRang+2)} for  i in range(1,NumberOfRows+1)} #ne sert qu'à construire la liste des places vides
for i in range(NbPassengers):
    places_vides_b[dict_attributions[i][0]][dict_attributions[i][1]] = False
    if passenger_sheet[i]['cat'] == 'WCHB':
        for a in range(-1,3):
            for b in range(-1,2):
                places_vides_b[dict_attributions[i][0]+a][dict_attributions[i][1]+b] = False
for i in places_vides_b.keys():
    if i<= NumberOfBusinessSeats//4:
        for j in places_vides_b[i].keys():
            places_vides_b[i][j]=False


places_vides = []
for i in places_vides_b.keys():
    for j in places_vides_b[i].keys():
        if places_vides_b[i][j] and j!=4 and i>NumberOfBusinessSeats//4:
            places_vides.append((i,j))


def mutation(solution, group_dict, passenger_sheet):
    """solution : [attributions, scores, candidats, places_vides]
    choice = 1 : Echange deux passagers échangeables, et garde la solution si le score d'un des groupes a été amélioré
    choice = 2 : Place un passager dans une place vide, et garde la solution si le score de son groupe a été amélioré"""
    attributions = solution[0]
    new_attr = {i:attributions[i] for i in attributions.keys()}
    scores = solution[1]
    new_scores = {i:scores[i] for i in scores.keys()}
    candidats = solution[2]
    vides = solution[3].copy()
    choice = random.randint(1,2)
    i2 = random.randint(0,len(candidats[1])-1)
    c2 = candidats[1][i2]
    if choice==1:
        i1 = random.randint(0,len(candidats[0])-1)
        c1 = candidats[0][i1]
        new_attr[c1], new_attr[c2]=attributions[c2], attributions[c1]
        gr1 = passenger_sheet[c1]['groupnr']
        gr2 = passenger_sheet[c2]['groupnr']
        new_scores[gr1] = score_groupe(gr1, group_dict, new_attr)
        new_scores[gr2] = score_groupe(gr2, group_dict, new_attr)
        if contrainte_barycentre(new_attr) and (new_scores[gr1]<scores[gr1] or new_scores[gr2]<scores[gr2]):
            new_scores[-1]+= (new_scores[gr1] - scores[gr1]) + (new_scores[gr2] - scores[gr2])
            return [new_attr, new_scores, candidats, vides]
        else:
            return new_scores[-1]
    else:
        id_vide = random.randint(0,len(vides)-1)
        el = vides[id_vide]
        vides.append(attributions[c2])
        new_attr[c2] = vides[id_vide]
        vides.remove(el)
        gr2 = passenger_sheet[c2]['groupnr']
        new_scores[gr2] = score_groupe(gr2, group_dict, new_attr)
        if contrainte_barycentre(new_attr) and (new_scores[gr2]<scores[gr2]):
            new_scores[-1]+= (new_scores[gr2] - scores[gr2])
            return [new_attr, new_scores, candidats, vides]
        else:
            return new_scores[-1]
    
def keep_n_best(n, generation):
    """ prend les n meilleurs élément d'une génération et renvoie la nouvelle génération d'élite"""
    tri = generation
    tri.sort(key= lambda sol: sol[1][-1])
    return tri[0:n]

    
def enfants(group_dict, solution, passenger_sheet, nb_enfants, limit):
    """Crée une nouvelle génération à partir d'un individu en lui appliquant une mutation"""
    new_gen = [solution]
    iter = 0
    while len(new_gen)<nb_enfants and iter<limit:
        mutated = mutation(solution, group_dict, passenger_sheet)
        if type(mutated)!=int:
            new_gen.append(mutated)
        iter+=1
    return new_gen

def heuristique(group_dict, dict_attributions, passenger_sheet, nb_enfants, limitc, limitg, limitf):
    """Crée des générations avant de sélectionner le meilleur individu pour la génération suivante;
    limitc : nombre maximum de tentatives de mutation par génération ;
    limitg : nombre maximum de génération ;
    limitf : nombre maximum de générations successives sans amélioration du score"""
    candidats = candidates(passenger_sheet, group_dict)
    scores = create_group_scores(group_dict, dict_attributions)
    print('Score initial : ' + str(scores[-1]))
    solution = [dict_attributions, scores, candidats, places_vides]
    gen = [solution]
    iter = 0
    best_score = gen[0][1][-1]
    failures = 0
    display_ = 1 #(variable utilisée pour l'affichage)
    while iter<limitg:
        iter += 1
        if int(display_*limitg/20)==iter:
            print('Génération numéro : ' + str(iter) + '  (' + str(5*display_) + '%)')
            display_+=1
        for sol in gen:
            gen = gen + enfants(group_dict, sol, passenger_sheet, nb_enfants, limitc)
        gen = keep_n_best(nb_enfants, gen) #on peut remplacer le 1 par nb_enfants
        temp_score = keep_n_best(1, gen)[0][1][-1]
        if temp_score==best_score:
            failures += 1
            print('fails : ' + str(failures) + '/'+str(limitf))
        else:
            failures = 0
        if failures == limitf:
            return keep_n_best(nb_enfants, gen)[0]
    print('--- LIMITE DE GENERATIONS ATTEINTE ---')
    return keep_n_best(nb_enfants, gen)[0]

            
```

```python
group_update = {i:True for i in group_dict}
for i in group_update:
    if (group_sheet[i]['WCHR']+group_sheet[i]['WCHB'])>0:
        group_update[i]=False


def permutliste(seq, er=False):
    """Retourne la liste de toutes les permutations de la liste seq (non récursif).
       si er=True: élimination des répétitions quand seq en a (ex: [1,2,2]
    """
    p = [seq]
    n = len(seq)
    for k in range(0,n-1):
        for i in range(0, len(p)):
            z = p[i][:]
            for c in range(0,n-k-1):
                z.append(z.pop(k))
                if er==False or (z not in p):
                    p.append(z[:])
    return p


def arrangliste(seq):
    """Liste des arrangements des objets de la liste seq pris k à k"""
    p = []
    i, imax = 0, 2**len(seq)-1
    while i<=imax:
        s = []
        j, jmax = 0, len(seq)-1
        while j<=jmax:
            if (i>>j)&1==1:
                s.append(seq[j])
            j += 1
        if len(s)==len(seq):
            v = permutliste(s)
            p.extend(v)
        i += 1 
    return p

# print(arrangliste([1,2,3,4]))


def echange_barycentre(id_sujet, id_cible, dict_reservations):
    gr1 = group_dict[id_sujet]
    gr2 = group_dict[id_cible]
    for gr1 in arrangliste(group_dict[id_sujet]):
        for gr2 in arrangliste(group_dict[id_cible]):
            temp_dict = {i:dict_reservations[i] for i in dict_reservations}
            for i in range(len(gr1)):
                temp_dict[gr1[i]],temp_dict[gr2[i]] = temp_dict[gr2[i]],temp_dict[gr1[i]]
            if contrainte_barycentre(temp_dict):
                return_dict = {}
                for i in range(len(gr1)):
                    return_dict[gr1[i]] = temp_dict[gr1[i]]
                    return_dict[gr2[i]] = temp_dict[gr2[i]]
                return return_dict



def placement_dynamique():
    dict_reservations = {i:result[0][i] for i in result[0]}
    ordre_reservation=list()
    for key in group_update:
        if group_update[key]:
            ordre_reservation.append(key)
    rd.shuffle(ordre_reservation)
    for id_group in ordre_reservation:
        list_choice = []
        for g in group_dict:
            if group_update[g] and (length_of_group(group_sheet[id_group]) == length_of_group(group_sheet[g])) and group_sheet[id_group]['class']==group_sheet[g]['class'] and ((group_sheet[g]['transittime']>0)==(group_sheet[id_group]['transittime']>0)):
                nouvelles_places = echange_barycentre(id_group,g, dict_reservations)
                if nouvelles_places:
                    list_choice.append(nouvelles_places) #nouvelles_places : {id_personne: nouvelle place for personne in (les deux groupes)}
        nv_id=rd.randint(0,len(list_choice)-1) #on choisit une possibilité au hasard ici, mais c'est l'étape où le groupe choisit ses places parmi les possibilités
        new_seats = list_choice[nv_id]
        print_seats = []
        for id_p in new_seats:
            print_seats.append(new_seats[id_p])
            dict_reservations[id_p] = new_seats[id_p]
        group_update[id_group] = False
        print('Le groupe ' + str(id_group) + ' a choisi les places :')
        print(print_seats)
    return dict_reservations
show = placement_dynamique()
```
